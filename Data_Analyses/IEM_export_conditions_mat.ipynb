```python
filenames = glob.glob('IEM_*.csv')
print filenames
```

```python
db = pd.DataFrame()
for name in filenames:
    id1 = filenames[0].find('_') + 1
    id2 = filenames[0].find('.')
    sub_nbr = name[id1:id2]
    s_id = 's_'+sub_nbr
    frame = pd.read_csv(name, header=0, index_col=False,names=colnames)
    frame['Subject'] = s_id
    frame['Trial'] = np.arange(36)
    db = pd.concat([db, frame], ignore_index=True)    
```
