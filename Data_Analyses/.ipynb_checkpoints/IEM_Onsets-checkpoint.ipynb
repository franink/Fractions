```python
filenames = glob.glob('IEM_*.csv')
print filenames
```

```python
db = pd.DataFrame()
for name in filenames:
    id1 = filenames[0].find('_') + 1
    id2 = filenames[0].find('.')
    sub_nbr = name[id1:id2]
    s_id = 's_'+sub_nbr
    frame = pd.read_csv(name, header=0, index_col=False,names=colnames)
    frame['Subject'] = s_id
    db = pd.concat([db, frame], ignore_index=True)    
```

```python
db.groupby('Subject')[['hit1','falseAlarm1','correctRejection1','miss1']].mean()
```

```python
db.drop(['Condition1','Condition2','Condition3','Condition4'], axis=1,inplace=True)
```

```python
db.rename(columns={'Conditon1':'Condition1','Conditon2':'Condition2','Conditon3':'Condition3','Conditon4':'Condition4'},
          inplace=True)
```

Setup some variables

```python
stim_dur = db.stim_t.loc[0]
ramp_up = db.ramp_up.loc[0]
```

'null' 1-4 indexes null trials that have to separate for onset creation

```python
runs = ['1','2','3','4']
subs = db.Subject.unique()
```

```python
nullstr = 'null'
onsetstr = 'stim_StartReal'
path = '/Users/frankanayet/Google Drive/NLineData/Beh_Data/NLine-Space/Space/OnsetFiles_LSS'
if not os.path.exists(path):
    os.makedirs(path)
for s in subs:
    for r in runs:
        run = 'run'+r
        nulltmp = nullstr+r
        onsettmp = onsetstr+r
        tmp = db[(db['Subject']==s) & (db[nulltmp] == 0)][onsettmp] - ramp_up
        tmp_db = pd.DataFrame.from_items([('time',tmp[tmp>=0]),('duration',stim_dur),('val',1)])
        tmp_path = '/Users/frankanayet/Google Drive/NLineData/Beh_Data/NLine-Space/Space/OnsetFiles_LSS_IEM/'+s+'/'
        if not os.path.exists(tmp_path):
            os.makedirs(tmp_path)
        tmp_filename = s+'_'+run+'_AllOnsets.txt'
        tmp_pathfile = tmp_path + tmp_filename
        tmp_db.to_csv(tmp_pathfile, sep='\t', header=False, index=False)
```
