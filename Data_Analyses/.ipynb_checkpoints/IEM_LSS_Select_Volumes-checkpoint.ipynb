```python
db = pd.DataFrame()
for name in filenames:
    id1 = filenames[0].find('_') + 1
    id2 = filenames[0].find('.')
    sub_nbr = name[id1:id2]
    s_id = 's_'+sub_nbr
    frame = pd.read_csv(name, header=0, index_col=False,names=colnames)
    frame['Subject'] = s_id
    frame['Trial'] = np.arange(36)
    db = pd.concat([db, frame], ignore_index=True)    
```

```python
nullstr = 'null'
condstr = 'Condition'
Xstr = 'stimLocsX'
Ystr = 'stimLocsY'
big_index_dict = {}
for s in subs:
    big_index_dict[s] = {}
    for val in values:
        big_index_dict[s][str(val)] = []
        for r in runs:
            run = 'run'+r
            nulltmp = nullstr+r
            condtmp = condstr+r
            trialtmp = trialstr + r
            Xtemp = Xstr+r
            Ytemp = Ystr+r
            tmp = db[(db['Subject']==s) & (db[condtmp] == val)][[Xtemp, Ytemp, trialtmp]]
            #tmp_trial = tmp.index.tolist()
            tmp_trial = tmp[trialtmp].values.astype(int)
            big_index_dict[s][str(val)].append(tmp_trial)
```
