```python
new_df['Line_pct_minusOne'] = np.nan
```

```python
new_df.Subject.unique()
```

```python
for s in new_df.Subject.unique():
    for r in new_df.Run.unique():
        for t in new_df.Task.unique():
            tmp = new_df.loc[new_df['Subject']==
                             s].loc[new_df['Run']==r].loc[new_df['Task']==t][['Line_pct', 'Trial']]
            new_tmp = np.empty(len(tmp))
            new_tmp[0] = np.nan
            for i in range(1,len(tmp)):
                new_tmp[i] = tmp['Line_pct'].iloc[i-1]
            new_tmp = pd.Series(new_tmp)
            new_df.loc[(new_df['Subject']==s) & (new_df['Run']==r) & (new_df['Task'] == t),
                       'Line_pct_minusOne'] = new_tmp.values        
```

```python
new_tmp
```

```python
new_df['Dist50_minusOne'] = abs(new_df['Line_pct_minusOne'] - 0.5)
```

```python
new_df[['Line_pct', 'Line_pct_minusOne', 'Dist50_minusOne']].head(40)
```

```python
new_df['quartile'] = new_df['Line_pct']
```

```python
new_df.loc[new_df['quartile'] < 0.25, 'quartile'] = 1
```

```python
new_df.loc[(new_df['quartile'] > 0.25) & (new_df['quartile'] < 0.5) , 'quartile'] = 2
```

```python
new_df.loc[(new_df['quartile'] > 0.5) & (new_df['quartile'] < 0.75) , 'quartile'] = 3
```

```python
new_df.loc[(new_df['quartile'] > 0.75) & (new_df['quartile'] < 1) , 'quartile'] = 4
```

```python
new_df['quartile'] = new_df['quartile'].astype('category')
```
